```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize an empty list to store comments
comment_list = []

# Launch Chrome browser
with Chrome() as driver:
    wait = WebDriverWait(driver, 10)
    driver.get(r"https://www.youtube.com/watch?v=tkbsPlJ8Amg")
    time.sleep(5)

    # Scroll down to load comments
    ActionChains(driver).send_keys(Keys.SPACE).perform()

    prev_comments_count = 0
    current_comments = driver.find_elements(By.CSS_SELECTOR, "#content-text")

    # Keep scrolling until no new comments are loaded
    while prev_comments_count < len(current_comments):
        prev_comments_count = len(current_comments)
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(3)
        current_comments = driver.find_elements(By.CSS_SELECTOR, "#content-text")

    # Append each comment to the list
    for comment in current_comments:
        comment_list.append(comment.text)

# Display the collected comments
for comment in comment_list:
    print(comment)
    print("-" * 80)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    text = text.lower()  # Convert text to lowercase
    text = " ".join([word for word in word_tokenize(text) if word.isalnum() and word not in stop_words])
    return text

# Example usage
cleaned_reviews = [preprocess_text(review) for review in comment_list]
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert the list of comments into a single string
corpus = ' '.join(comment_list)

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform([corpus])

# Get feature names and TF-IDF values
feature_names = tfidf_vectorizer.get_feature_names_out()
tfidf_values = tfidf_matrix.toarray()[0]

# Create a DataFrame for better visualization
tfidf_df = pd.DataFrame({'Word': feature_names, 'TF-IDF Score': tfidf_values})

# Print the DataFrame
print(tfidf_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(comment_list)

# K-means Clustering
num_clusters = 5  # You can adjust the number of clusters as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(tfidf_matrix)

# Get cluster labels for each comment
cluster_labels = kmeans.labels_

# Print the cluster assignments for each comment
for i, label in enumerate(cluster_labels):
    print("Comment {}: Cluster {}".format(i+1, label))

# Evaluate the clustering using silhouette score
silhouette_avg = silhouette_score(tfidf_matrix, cluster_labels)
print("\nSilhouette Score:", silhouette_avg)
```

```python
def print_cluster_info(tfidf_matrix, cluster_labels, comment_list, num_clusters):
    print("Comments in each cluster:")
    for cluster_num in range(num_clusters):
        print("\nCluster {}:".format(cluster_num))
        cluster_comments = [comment_list[i] for i, cluster in enumerate(cluster_labels) if cluster == cluster_num]
        for comment in cluster_comments[:3]:  # Print only the first 3 comments for each cluster
            print("-", comment)

# Example usage
print_cluster_info(tfidf_matrix, cluster_labels, comment_list, num_clusters)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def text_summarization(review_list, num_sentences=3):
    # Convert the list of reviews into a single string
    concatenated_text = ' '.join(review_list)
    
    # Tokenize the text into sentences
    sentences = concatenated_text.split('.')
    
    # Create a CountVectorizer to convert text into a matrix of token counts
    vectorizer = CountVectorizer(stop_words='english')
    X = vectorizer.fit_transform(sentences)
    
    # Compute cosine similarity matrix
    similarity_matrix = cosine_similarity(X, X)
    
    # Initialize sentence scores
    scores = np.zeros(len(sentences))
    
    # Iterate to calculate scores using PageRank algorithm
    damping_factor = 0.85
    for _ in range(10):
        for i in range(len(sentences)):
            scores[i] = (1 - damping_factor) + damping_factor * np.sum(similarity_matrix[i] * scores)
    
    # Get indices of top-ranked sentences
    top_indices = scores.argsort()[-num_sentences:][::-1]
    top_indices.sort()
    
    # Generate summary using top-ranked sentences
    summary = ' '.join([sentences[i] for i in top_indices])
    
    return summary

youtube_reviews_summary = text_summarization(comment_list)
print(youtube_reviews_summary)
```
